```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
source_dir = "./source/Fluxes/"
```

```python
test_bnb_data = np.genfromtxt(source_dir+'bnb_flux_file.csv', delimiter=',', skip_header=1)
bnb_data = np.genfromtxt(source_dir+'MicroBooNE_BNB_fullflux.txt', skip_header=1)

test_nue = test_bnb_data[:,1]
test_numu = test_bnb_data[:,2]
test_nuebar = test_bnb_data[:,3]
test_numubar = test_bnb_data[:,4]


energy = bnb_data[:len(test_bnb_data),0]
bnb_nue = bnb_data[:len(test_bnb_data),1]
bnb_numu = bnb_data[:len(test_bnb_data),2]
bnb_nuebar = bnb_data[:len(test_bnb_data),4]
bnb_numubar = bnb_data[:len(test_bnb_data),5]


np.savetxt(source_dir+'bnb_fullflux.txt', test_bnb_data, delimiter='    ', header='Energy   [GeV]    nue     numu    nuebar  numubar')
```

```python
scale = 1/0.05 # plot is scaled by GeV, flux file scaled by bin (0.05 GeV): "scaled to nu/POT/bin/cm2"


plt.step(energy, test_nue, 'r', label='nue test',where="mid",linewidth=1)
plt.step(energy, test_numu, 'g', label='numu test',where="mid",linewidth=1)
plt.step(energy, test_nuebar, 'b', label='nuebar test',where="mid",linewidth=1)
plt.step(energy, test_numubar, 'y', label='numubar test',where="mid",linewidth=1)


plt.step(energy, scale*bnb_nue, 'r--', label='Nue',where="mid",linewidth=1)
plt.step(energy, scale*bnb_numu, 'g--', label='Numu',where="mid",linewidth=1)
plt.step(energy, scale*bnb_nuebar, 'b--', label='Nuebar',where="mid",linewidth=1)
plt.step(energy, scale*bnb_numubar, 'y--', label='Numubar',where="mid",linewidth=1)
plt.legend()
plt.yscale('log')

plt.xlabel('True neutrino energy [GeV]')
plt.ylabel(r'$\nu$/cm$^2$/POT/GeV')
```

```python
numi_cc = np.genfromtxt(source_dir+'numi_numu_fhc_cc_cs.csv', delimiter=',', skip_header=0)
numi_miller = np.genfromtxt(source_dir+'numi_fhc_miller.csv', delimiter=',', skip_header=1)

nue = numi_miller[:,1]
nuebar = numi_miller[:,2]
numu = numi_miller[:,3]
numubar = numi_miller[:,4]

np.savetxt(source_dir+'numi_fhc_fullflux.txt', numi_miller, delimiter='    ', header='Energy   [GeV]    nue     numu    nuebar  numubar')
```

```python

scale_cc = 1/(0.05*2.4*10**20)  # 0.05 GeV/bin, 2.4e20 POT, 1/cm2
plt.step(numi_cc[:,0], scale_cc*numi_cc[:,1], 'r', label='numu cc',where="mid",linewidth=1)

scale_miller = 1
plt.step(numi_miller[:,0], scale_miller*nue, 'g', label='nue',where="mid",linewidth=1)
plt.step(numi_miller[:,0], scale_miller*nuebar, 'b', label='nuebar',where="mid",linewidth=1)
plt.step(numi_miller[:,0], scale_miller*numu, 'y', label='numu',where="mid",linewidth=1)
plt.step(numi_miller[:,0], scale_miller*numubar, 'c', label='numubar',where="mid",linewidth=1)

plt.yscale('log')
plt.legend()
plt.xlabel('True neutrino energy [GeV]')
plt.ylabel(r'$\nu$/cm$^2$/POT/GeV')
```
